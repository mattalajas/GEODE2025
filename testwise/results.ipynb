```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: [min_A, max_A, min_B, max_B] per model
metric_array = np.array([
    [16.6,22.28,18.91,34.15],
    [24.22,31.14,27.58,35.23],
    [33.33,38.49,33.16,42.36],
    [29.67,44.57,37.1,40.43],
    [23.65,35.11,21.67,27.16],
    [25.06,32.64,26.09,33.32],
    [24.58,33.32,23.24,35.34],
    [29.91,40.27,31.42,38.72],
    [14.63,18.33,16.21,22.45],
    [14.09,17.96,15.66,21.81]
])

n_models = metric_array.shape[0]

# Prepare tidy DataFrame for seaborn-style grouping
data = []

for i in range(n_models):
    model = f"Model {i+1}"
    # Metric A
    data.append({"Model": model, "Metric": "A", "Part": "Min", "Value": metric_array[i, 0]})
    data.append({"Model": model, "Metric": "A", "Part": "Gap", "Value": metric_array[i, 1] - metric_array[i, 0]})
    # Metric B
    data.append({"Model": model, "Metric": "B", "Part": "Min", "Value": metric_array[i, 2]})
    data.append({"Model": model, "Metric": "B", "Part": "Gap", "Value": metric_array[i, 3] - metric_array[i, 2]})

df = pd.DataFrame(data)

# Pivot to make stacking easier
pivot_df = df.pivot_table(index=["Model", "Metric"], columns="Part", values="Value").reset_index()

# Set up the bar plot
fig, ax = plt.subplots(figsize=(8, 5))

# Create a color palette
colors = {
    'A': ('lightblue', 'steelblue'),
    'B': ('wheat', 'orange')
}

# Determine bar positions
model_order = df["Model"].unique()
x = np.arange(len(model_order))
bar_width = 0.35

for i, model in enumerate(model_order):
    for j, metric in enumerate(["A", "B"]):
        row = pivot_df[(pivot_df["Model"] == model) & (pivot_df["Metric"] == metric)]
        if not row.empty:
            min_val = row["Min"].values[0]
            gap_val = row["Gap"].values[0]
            x_pos = i + (j - 0.5) * bar_width  # left/right offset for A/B
            base_color, top_color = colors[metric]

            ax.bar(x_pos, min_val, width=bar_width, color=base_color, label=f"{metric} Min" if i == 0 else "")
            ax.bar(x_pos, gap_val, width=bar_width, bottom=min_val, color=top_color, label=f"{metric} Max−Min" if i == 0 else "")

# Final plot touches
ax.set_xticks(x)
ax.set_xticklabels(model_order)
ax.set_ylabel("Performance")
ax.set_title("Side-by-Side Stacked Bars per Model using Seaborn Grouping Style")
ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
sns.despine()

plt.tight_layout()
plt.show()
```
